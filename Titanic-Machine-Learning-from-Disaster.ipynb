```python
#import Libraries
import pandas
import tensorflow
import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
import tensorflow as tf
```

```python
df = pandas.read_csv('train.csv') #Converting into Dataframe
```

```python
df.head()
```

```python
df['Sex'] = [0 if i=='male' else 1 for i in df['Sex']] #Converting into binary
```

```python
x_train = df[['PassengerId','Pclass','Sex']] #Feature engineering, This will be our input
```

```python
y = to_categorical(df['Survived']) # Encoding the labels 
```

```python
#Using class for the accuracy of our model
class Callback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('acc')>0.80):
            print('Stop training Your model is trained!')
            self.model.stop_training = True
```

```python
callbacks = Callback()
```

```python
#making model 
model = Sequential()
model.add(Dense(28, activation='relu', input_shape=(3,)))
model.add(Dense(2, activation='softmax'))
```

```python
#compile our model with an optimizer , loss_function and checking accuracy 
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

```python
# Getting fit our model for training 
model.fit(x_train, y, epochs=4000, callbacks=[callbacks])
```
